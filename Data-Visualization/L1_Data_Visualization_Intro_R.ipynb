**Activity 3:** Find the correlation between the variables in the dataset and find the pairs which have correlation more than 0.5. 

*Hint: there are 4 pairs of them.*

Programmably **list them** or **create a visualization** that one can read the answer off of it easily, as opposed to just printing a correlation matrix.

```r
## Your answer for activity 3 goes here ---
library(dplyr)
library(tidyr)

USDA_cor =
cor(USDA) %>%
  as.data.frame() %>%
  mutate(var1 = rownames(.)) %>%
  gather(var2, value, -var1) %>%
  arrange(desc(value)) %>%
  group_by(value) %>%
  filter(row_number()==1)


USDA_cor2 = subset(USDA_cor, value > .5 & value < 1)
USDA_cor2

```

**Activity 4:** Fit a linear regression model between Calories and TotalFat.

```r
## Your answer for activity 4 goes here ---
library(ggplot2)

lm(Calories ~ TotalFat, data = USDA)

p1 <- ggplot(USDA, aes(x=Calories, y=TotalFat)) + geom_point() 
p1 <- p1 + geom_smooth(method = lm) + expand_limits(x = 4, y = 4)
p1
```
