## Step 4: Vet Data Sources

Use the cells below to load the data, inspect it, and perform some exploratory data analysis.  

Add more cells (`Insert > Insert Cell Below`) if you want additional cells.

```r
library(dplyr)
library(ggplot2)
library(RColorBrewer)

mls = read.csv('mls_2019_stats.csv')

# replace FC Cincinnati NA values with mean values of column, data unavailable due to new team
mls$Revenue[which(is.na(mls$Revenue))] = mean(mls$Revenue,na.rm = TRUE)
mls$Operating.Income[which(is.na(mls$Operating.Income))] = mean(mls$Operating.Income,na.rm = TRUE)

mls

mls2 = read.csv('Salary-List-September-2019.csv')
mls2 <- subset(mls2, select = -c(X))
head(mls2)
```

## Step 5: Filter Results

Perform you data carpentry that helps in your story telling here.

Add more cells (`Insert > Insert Cell Below`) if you want additional cells.

```r
# Top 20 highest paid players during the mls season

head(mls2[order(-mls2$Salary),], 20)
```

```r
# Average salary of each 2019 mls team

var=
mls2 %>%
group_by(Club)%>%
summarise(avg=mean(Salary)) %>% as.data.frame
names(var)[names(var) == "avg"] = "Salary.Average"

# var[order(-var$Salary.Average),]

df = merge(mls, var, by="Club") 
df
```

```r
# finding the average guaranteed compensation per team for the 2019 season.

var2=
mls2 %>%
group_by(Club)%>%
summarise(avg=mean(Guaranteed.Comp)) %>% as.data.frame

names(var2)[names(var2) == "avg"] = "Guaranteed.Comp.Average"

# var2[order(-var2$Guaranteed.Comp.Average),]

final.mls = merge(df, var2, by="Club") 
head(final.mls)
```

```r
# Find the highest correlation with Wins

cor(final.mls$Value, final.mls$Wins)
cor(final.mls$Revenue, final.mls$Wins)
cor(final.mls$Operating.Income, final.mls$Wins)
cor(final.mls$Salary.Average, final.mls$Wins)
cor(final.mls$Guaranteed.Comp.Average, final.mls$Wins)
```

According to the outputs of the correlations the Valuation of a team has the best correlation with Wins.Meaning it could be a good indicator of how likely your team will win games during the season. The second strongest category is Revenue, which would make sense as this is what a team can spend on players salaries and other things to improve the team.

```r
# Correlation Matrix : You can see all the relationships of the different variables
nums = sapply(final.mls, is.numeric)
cor(final.mls[ , nums])
```

## Step 6: Visuals 

Build up your key visual story elements!

Add more cells (`Insert > Insert Cell Below`) if you want additional cells.

I will be plotting the two highest correlations with Wins to visualize the realtionships.
I chose to focus on wins because this is the best indicator of how a team is performing.

```r
plot(final.mls$Wins, final.mls$Value)
plot(final.mls$Wins, final.mls$Salary.Average)
```

```r
# Plotting the relationship between Value and Wins, you can start to see the correlation.

p <- ggplot(final.mls, aes(x=Wins,y=Value)) + geom_point(alpha=0.5, color="Blue")
p
```

```r
# Plotting the correlation between Value and Wins per team with best fitted line.

ggplot(final.mls, aes(x = Wins, y = Value, colour = Club))  + 
    geom_smooth(aes(color = "Red"), method = "lm") + geom_point()
```

```r
# Plotting the relationship between Salary.Average and Wins, you can start to see the correlation.

p <- ggplot(final.mls, aes(x=Wins,y=Salary.Average)) + geom_point(alpha=0.5, color="Blue")
p
```

```r
#Plotting the correlation between Average Salary and Wins per team with best fitted line.

ggplot(final.mls, aes(x = Wins, y = Salary.Average, colour = Club))  + 
    geom_smooth(aes(color = "Red"), method = "lm") + geom_point()
```

```r
boxplot(Value~Wins,data=final.mls, main="Wins vs Valuation",
   xlab="Wins", ylab="Value", col="orange")
```

```r
boxplot(Salary.Average~Wins,data=final.mls, main="Wins vs Average Salary",
   xlab="Wins", ylab="Salary.Average", col="purple")
```

```r
# lastly I am adding in a size factor into my plots.

p <- ggplot(final.mls, aes(x=Wins,y=Value, size = Salary.Average)) + geom_point(alpha=0.5, color="Orange")
p

p <- ggplot(final.mls, aes(x=Wins,y=Salary.Average, size = Value)) + geom_point(alpha=0.5, color="Purple")
p
```
